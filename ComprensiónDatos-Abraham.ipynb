BloodPreassure: Se refiere al indicador de presion en la sangre de cada persona. Es una variable cuantitativa (discreta). Se ecuentra en los rangos de 0 y 199.  
SkinThickness: Se refiere al indicador de grosor de la piel de cada persona. Es una variable cuantitativa (discreta). Se ecuentra en los rangos de 0 y 99.

DiabetesPedigreeFunction: Se refiere a la sÃ­ntesis de la historia de diabetes mellitus en familiares y la relaciÃ³n genÃ©tica de esos familiares con el sujeto. Es una variable cuantitativa (discreta). Se ecuentra en los rangos de 2.42 y 0.078.  
Age: Se refiere a la edad de cada persona. Es una variable cuantitativa (discreta). Se ecuentra en los rangos de 21 y 81.  
Outcome: Se refiere a si una persona tiene diabetes o no. Es una variable cualitativa ordinal. Toma valores de 0 y 1.  

Basandonos en la media, mediana y desviacion estandar de cada variable se podria generalizar concluyendo que se tiene una alta desviacion estandar. Para ver si se pueden hacer predicciones sobre la diabetes en las personas habria que sacar que tanto se ajustan los datos a la linea.
